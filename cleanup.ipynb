```python
import pandas as pd

import numpy as np

import os
```

```python
file_path = "layoffs - Copy.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
#drop funds raised

df.drop(['funds_raised'], axis=1)
```

```python
df['country'].value_counts()
```

```python
# dropping stage column as well

df.drop(['stage'], axis=1)
```

```python
# convert dates to datetime

df['date']= pd.to_datetime(df['date'])
print(df.dtypes)
```

```python
# sort by quarter one in 2022


df2= df[(df['date']>"2021-12-31")&(df['date']<"2022-03-31")]
```

```python
df2.tail()
```

```python
# dropping stage and funds_raised columns df2

df3= df2.drop(['stage', 'funds_raised'], axis=1)
```

```python
df3.tail()
```

```python
# all countries and nan values for 1st qtr 2022
df3.head()
```

```python
# dropping nan's 

df_1st= df3.dropna()

print(df_1st)
```

```python
# counting how many values we have for 1st qtr after dropping nan's

df_1st.count()
```

```python
df_4= df[(df['date']>"2022-09-30")&(df['date']<"2023-01-01")]

df_4.count()
```

```python
# dropping columns for 4th qtr

df_4.drop(['stage', 'funds_raised'], axis=1)
```

```python
# dropping nan values from 4th qtr

df_4th= df_4.dropna()
print(df_4th)
```

```python
# 4th qtr count 
df_4th.count()
```

```python
# create csv for qtr 1 cleaned

df_1st.to_csv("layoffClean1st.csv", encoding="utf-8", index = False)
```

```python
# create csv for qtr 4 cleaned

df_4th.to_csv("layoffClean4th.csv", encoding="utf-8", index = False)
```
