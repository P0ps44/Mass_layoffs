```python
import pandas as pd

import numpy as np

import os
```

```python
file_path = "layoffs - Copy.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
# add a new column with ID

df= df.assign(ID=range(len(df)))
df.tail()
```

```python
df['date']= pd.to_datetime(df['date'])
print(df.dtypes)
```

```python
# dropping two columns stage and funds_raised

df2= df.drop(['stage', 'funds_raised'], axis=1)

df2.head()
```

```python
# recreate qtr one again

qtr1_df= df2[(df['date']>"2021-12-31")&(df['date']<"2022-03-31")]

qtr1_df.head()
```

```python
# dropping nan's 

qtr1_df_clean= qtr1_df.dropna()

qtr1_df_clean.head()
```
